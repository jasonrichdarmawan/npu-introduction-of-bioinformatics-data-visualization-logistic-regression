```python
# Load GSE2034
df_gse2034 = pd.read_csv(
  "./dataset/GSE2034_series_matrix.txt",
  sep="\t",
  skiprows=54
)
df_gse2034
```

```python
# Load GPL96
df_gpl96 = pd.read_csv(
  "./dataset/GPL96-57554.txt",
  sep="\t",
  skiprows=16
)
df_gpl96
```
